```python
# The Melbourne data has some missing values (some houses for which some variables weren't recorded.)
# We'll learn to handle missing values in a later tutorial.  
# Your Iowa data doesn't have missing values in the columns you use. 
# So we will take the simplest option for now, and drop houses from our data. 
# Don't worry about this much for now, though the code is:

# dropna drops missing values (think of na as "not available")
#melbourne_data = melbourne_data.dropna(axis=0)
```

```python
X = cd[['Rooms','Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude',
       'Longtitude', 'Propertycount']]
```

```python
y = cd['Price']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
X = std.fit_transform(X)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(presort=True,splitter='random')
```

```python
dtr.fit(X_train,y_train)
```

```python
dtr_pred = dtr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print('MAE:',mean_absolute_error(y_test,dtr_pred))
print('MSE:',mean_squared_error(y_test,dtr_pred))
print('RMSE:',np.sqrt(mean_squared_error(y_test,dtr_pred)))
```

```python
print(dtr_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=200,random_state=0,n_jobs=3)
```

```python
rfr.fit(X_train,y_train)
```

```python
rfr_pred = rfr.predict(X_test)
```
